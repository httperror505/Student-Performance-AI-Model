```python
import pandas as pd
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import accuracy_score

# Load the CSV data file
# student_dataset = pd.read_csv("/content/sample_data/cleandiguess.csv")
student_dataset = pd.read_csv("/content/partially_cleaned.csv")  # Ensure you've uploaded the file correctly


# Check the first few rows of the data
print("Initial Dataset Head:")
print(student_dataset.head())

# Encode categorical variables
number = LabelEncoder()
categorical_columns = ['Gender', 'ParentalEducation', 'ParentalSupport',
                       'Extracurricular', 'GradeClass']
for column in categorical_columns:
    student_dataset[column] = number.fit_transform(student_dataset[column])

# Define feature columns and target variable
features = ['Age', 'Gender', 'ParentalEducation', 'StudyTimeWeekly',
            'Absences', 'Tutoring', 'ParentalSupport', 'Extracurricular']
target = 'GradeClass'

# Standardize numerical features for Gaussian Naive Bayes
scaler = StandardScaler()
student_dataset[features] = scaler.fit_transform(student_dataset[features])

# Split the data into training and testing sets (80% training, 20% testing)
features_train, features_test, target_train, target_test = train_test_split(
    student_dataset[features], student_dataset[target], test_size=0.20, random_state=42
)

# Initialize Gaussian Naive Bayes model
model = GaussianNB()

# Train the model
model.fit(features_train, target_train)

# Perform cross-validation for better accuracy estimation
cross_val_accuracies = cross_val_score(model, student_dataset[features], student_dataset[target], cv=5)
print(f"Cross-Validation Accuracies: {cross_val_accuracies}")
print(f"Mean Cross-Validation Accuracy: {cross_val_accuracies.mean() * 100:.2f}%")

# Make predictions on the test data
pred = model.predict(features_test)

# Calculate accuracy on test data
accuracy = accuracy_score(target_test, pred)
print(f"Test Set Accuracy: {accuracy * 100:.2f}%")

# Make a prediction with a new sample (example input: standardized values)
new_sample = scaler.transform([[17, 1, 2, 4, 2, 1, 2, 0,]])
predicted_class = model.predict(new_sample)

# Interpret the prediction
grade_classes = {
    0: 'A (GPA >= 3.5)',
    1: 'B (3.0 <= GPA < 3.5)',
    2: 'C (2.5 <= GPA < 3.0)',
    3: 'D (2.0 <= GPA < 2.5)',
    4: 'F (GPA < 2.0)'
}
print(f"\nPredicted GradeClass: {grade_classes[predicted_class[0]]} : {predicted_class}")
from sklearn.metrics import confusion_matrix, classification_report, precision_score, recall_score, f1_score

# Confusion Matrix
conf_matrix = confusion_matrix(target_test, pred)
print("Confusion Matrix:")
print(conf_matrix)

# Classification Report (Precision, Recall, F1-score for each class)
print("\nClassification Report:")
print(classification_report(target_test, pred, target_names=grade_classes.values()))

# Calculate precision, recall, and F1 score for each class (macro, weighted, etc.)
precision = precision_score(target_test, pred, average='weighted')  # Use 'macro' for unweighted mean across classes
recall = recall_score(target_test, pred, average='weighted')
f1 = f1_score(target_test, pred, average='weighted')

print(f"\nOverall Metrics:")
print(f"Precision (Weighted): {precision:.2f}")
print(f"Recall (Weighted): {recall:.2f}")
print(f"F1 Score (Weighted): {f1:.2f}")
```

```python
import pandas as pd
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import accuracy_score

# Load the CSV data file
student_dataset = pd.read_csv("/content/partially_cleaned.csv")

# Check the first few rows of the data
print("Initial Dataset Head:")
print(student_dataset.head())

# Encode categorical variables
number = LabelEncoder()
categorical_columns = ['Gender', 'ParentalEducation', 'ParentalSupport',
                       'Extracurricular', 'Sports', 'Music', 'Volunteering', 'GradeClass']
for column in categorical_columns:
    student_dataset[column] = number.fit_transform(student_dataset[column])

# Create a new derived feature: ActiveParticipation (logical OR of Extracurricular and Volunteering)
student_dataset['ActiveParticipation'] = (
    student_dataset['Extracurricular'] | student_dataset['Volunteering']
)

# Define feature columns (include the new feature) and target variable
features = ['Age', 'Gender', 'ParentalEducation', 'StudyTimeWeekly',
            'Absences', 'Tutoring', 'ParentalSupport', 'Extracurricular',
            'Sports', 'Music', 'Volunteering', 'ActiveParticipation']
target = 'GradeClass'

# Standardize numerical features for Gaussian Naive Bayes
scaler = StandardScaler()
student_dataset[features] = scaler.fit_transform(student_dataset[features])

# Split the data into training and testing sets (80% training, 20% testing)
features_train, features_test, target_train, target_test = train_test_split(
    student_dataset[features], student_dataset[target], test_size=0.20, random_state=42
)

# Initialize Gaussian Naive Bayes model
model = GaussianNB()

# Train the model
model.fit(features_train, target_train)

# Perform cross-validation for better accuracy estimation
cross_val_accuracies = cross_val_score(model, student_dataset[features], student_dataset[target], cv=5)
print(f"Cross-Validation Accuracies: {cross_val_accuracies}")
print(f"Mean Cross-Validation Accuracy: {cross_val_accuracies.mean() * 100:.2f}%")

# Make predictions on the test data
pred = model.predict(features_test)

# Calculate accuracy on test data
accuracy = accuracy_score(target_test, pred)
print(f"Test Set Accuracy: {accuracy * 100}%")

# Make a prediction with a new sample (example input: standardized values)
new_sample = scaler.transform([[17, 1, 2, 4, 2, 1, 2, 0, 0, 1, 0, 1]])  # Include ActiveParticipation
predicted_class = model.predict(new_sample)

# Interpret the prediction
grade_classes = {
    0: 'A (GPA >= 3.5)',
    1: 'B (3.0 <= GPA < 3.5)',
    2: 'C (2.5 <= GPA < 3.0)',
    3: 'D (2.0 <= GPA < 2.5)',
    4: 'F (GPA < 2.0)'
}
print(f"\nPredicted GradeClass: {grade_classes[predicted_class[0]]}")
```
